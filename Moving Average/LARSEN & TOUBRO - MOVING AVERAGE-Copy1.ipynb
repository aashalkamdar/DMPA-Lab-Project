```python
df = pd.read_csv('data/l_and_t/larsen_and_toubro.csv')
lt = pdr.DataReader('LT.BO', data_source='yahoo', start='2015-10-23', end='2019-12-30')
```

```python
# visualize the adj. closing price history
plt.title('Adjusted Close Price History')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Adjusted Close Price in Rupee', fontsize=18)
plt.plot(lt['Adj Close'])
plt.figure(figsize=(20,10))
plt.show()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.info()
```

```python
# setting the index as date
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
df.index = df['Date']
```

```python
#creating dataframe with date and the target variable
data = df.sort_index(ascending=True, axis=0)
new_data = pd.DataFrame(index=range(0,len(df)),columns=['Date', 'Adj Close'])
```

```python

for i in range(0,len(data)):
     new_data['Date'][i] = data['Date'][i]
     new_data['Adj Close'][i] = data['Adj Close'][i]
```

```python
train = new_data[:924]
valid = new_data[924:]
```

```python
# shapes of training set
print('\n Shape of training set:')
print(train.shape)

# shapes of validation set
print('\n Shape of validation set:')
print(valid.shape)
```

```python
preds = []
for i in range(0,valid.shape[0]):
    a = train['Adj Close'][len(train)-103+i:].sum() + sum(preds)
    b = a/103
    preds.append(b)
```

```python
# checking the results (RMSE value)
rms=np.sqrt(np.mean(np.power((np.array(valid['Adj Close'])-preds),2)))
print('\n RMSE value on validation set:')
print(rms)
```

```python
#plot
valid['Predictions'] = 0
valid['Predictions'] = preds
plt.plot(train['Adj Close'])
plt.plot(valid[['Adj Close', 'Predictions']])
```

```python
import math
from sklearn.metrics import mean_squared_error

math.sqrt(mean_squared_error(valid['Adj Close'],preds))
# rms=np.sqrt(np.mean(np.power((np.array(valid['Close'])-preds),2)))
```
